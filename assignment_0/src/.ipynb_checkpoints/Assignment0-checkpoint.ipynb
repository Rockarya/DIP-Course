```python
# Importing all needed libraries
import cv2
import matplotlib.pyplot as plt
```

```python
# loading all 4 images using opencv
img1 = cv2.imread("../images/img1.jpeg")
img2 = cv2.imread("../images/img2.jpeg")
img3 = cv2.imread("../images/img3.jpeg")
img4 = cv2.imread("../images/img4.jpeg")
```

```python
# Comment below 4 lines and see the effect of not converting OpenCV's imread BGR ordering to RGB.
# Here we are converting BGR ordering to RGB ordering
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)

# Displaying the images using matplotlib (making a 2x2 grid using subplots)
# create figure
fig = plt.figure(figsize=(10, 10))

# setting values to rows and column variables(can change to any m*n plot u want)
rows = 2
columns = 2

# Adding subplot for position 1
fig.add_subplot(rows, columns, 1)

# Displaying image1
plt.axis('off')
plt.title("Atif Aslam")
plt.imshow(img1)

# Adding subplot for position 2
fig.add_subplot(rows, columns, 2)

# Displaying image2
plt.axis('off')
plt.title("Akshay Kumar")
plt.imshow(img2)

# Adding subplot for position 3
fig.add_subplot(rows, columns, 3)

# Displaying image3
plt.axis('off')
plt.title("Kane Williamson")
plt.imshow(img3)

# Adding subplot for position 4
fig.add_subplot(rows, columns, 4)

# Displaying image4
plt.axis('off')
plt.title("Forest Road")
plt.imshow(img4)
```

# Why the above images seems unusual when the first 4 lines of above code are commented?

The reason for images being unusual is because OpenCV's function imread converts the image into array where the ordering is BGR instead of RGB, whereas most of the image processing libraries use RGB ordering. matplotlib library itself uses RGB ordering for image processing, hence we are changing BGR ordering into RGB ordering in first 4 lines of above code.

```python
# Converting colour image 2(Akshay Kumar) to grayscale using converion formula and the matplotlib library
import matplotlib.image as mpimg

colored_img = mpimg.imread("../images/img2.jpeg")

R, G, B = colored_img[:,:,0], colored_img[:,:,1], colored_img[:,:,2]
gray_img = 0.2989 * R + 0.5870 * G + 0.1140 * B
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.title("Akshay Kumar in grayscale")
plt.show()

# saving image in "images" folder
cv2.imwrite('../images/akshay_kumar_grayscale.jpeg',gray_img)
```

```python
# printing the dimensions of the 4th image(stored as img4)
print("Rows, Columns, Channels:",img4.shape)
```

```python
# Printing the range of values in the image4.
print('Maximum RGB value in the image4 is: {}'.format(img4.max()))
print('Minimum RGB value in the image4 is: {}'.format(img4.min()))
```
