```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import time
```

```python
def colored_to_gray(img):
    R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
    gray_img = 0.2989 * R + 0.5870 * G + 0.1140 * B
    return gray_img
```

```python
#1.1
def simple_mean_filter(img,k):
    start = time.time()
    h, w = img.shape
    p = (k-1)//2
    new_img = np.zeros((h, w), img.dtype)
    for x in range(w):
        for y in range(h):
            sum = 0
            for i in range(-p,p+1):
                for j in range(-p,p+1):
                    if x+i>=0 and x+i<w and y+j>=0 and y+j<h:
                        sum+=img[x+i,y+j]
            new_img[x,y]=(sum//(k*k))
            
    end = time.time()
    return end-start,new_img


img = cv2.imread('../images/ceramic.jpg')
img = colored_to_gray(img)
plt.imshow(img, cmap='gray')
plt.title('original image')
plt.xticks([]), plt.yticks([]) 
plt.show()

for k in range(3,14,2):
    t, op_img = simple_mean_filter(img,k)
    plt.imshow(op_img,cmap='gray')
    plt.title('grayscale ceramic image with {}*{} simple mean filter'.format(k,k))
    plt.xticks([]), plt.yticks([]) 
    plt.show()
```

```python
# 1.2
# instead of calculating sum over and over again we will use a cumulative sum array which will give sum for window of size k
def optimised_mean_filter(img,k,dp):
    start = time.time()
    h, w = img.shape
    new_img = np.zeros((h, w), img.dtype)
    p = (k-1)//2
    for x in range(1,w+1):
        for y in range(1,h+1):
            new_img[x-1,y-1] = (dp[min(w,x+p)][min(h,y+p)] - dp[max(0,x-p-1)][min(h,y+p)] - dp[min(w,x+p)][max(0,y-p-1)] + dp[max(0,x-p-1)][max(0,y-p-1)])//(k*k)
      
    end = time.time()
    return end-start,new_img

plt.imshow(img, cmap='gray')
plt.title('original image')
plt.xticks([]), plt.yticks([]) 
plt.show()
    
# calculating the cumulative array, to be used for optimised computation
h, w = img.shape
dp = np.zeros((h+1, w+1), img.dtype)

for x in range(1,w+1):
    for y in range(1,h+1):
        dp[x][y] = img[x-1,y-1] + dp[x-1][y] + dp[x][y-1] - dp[x-1][y-1]
    
        

for k in range(3,14,2):
    t, op_img = optimised_mean_filter(img,k,dp)
    plt.imshow(op_img,cmap='gray')
    plt.title('grayscale ceramic image with {}*{} optimised mean filter'.format(k,k))
    plt.xticks([]), plt.yticks([]) 
    plt.show()
```

```python
# 1.3
x = [3, 5, 7, 9, 11, 13]
dim = 8
while dim<=512:
    new_img = cv2.resize(img, (dim,dim))
    
    dim_efficient = []
    dim_bruteforce = []
    for k in range(3,14,2):
        t, op_img = optimised_mean_filter(new_img,k,dp)
        dim_efficient.append(t)
        
        t, op_img = simple_mean_filter(new_img,k)
        dim_bruteforce.append(t)
        
    if dim == 8:
        plt.plot(x,dim_efficient, color = 'blue', label = 'efficient version on 8*8 image')
        plt.plot(x,dim_bruteforce, color = 'blue', label = 'bruteforce version on 8*8 image', linestyle='dashed')
    elif dim == 16:
        plt.plot(x,dim_efficient, color = 'green', label = 'efficient version on 16*16 image')
        plt.plot(x,dim_bruteforce, color = 'green', label = 'bruteforce version on 16*16 image', linestyle='dashed') 
    elif dim == 32:
        plt.plot(x,dim_efficient, color = 'red', label = 'efficient version on 32*32 image')
        plt.plot(x,dim_bruteforce, color = 'red', label = 'bruteforce version on 32*32 image', linestyle='dashed')
    elif dim == 64:
        plt.plot(x,dim_efficient, color = 'cyan', label = 'efficient version on 64*64 image')
        plt.plot(x,dim_bruteforce, color = 'cyan', label = 'bruteforce version on 64*64 image', linestyle='dashed')
    elif dim == 128:
        plt.plot(x,dim_efficient, color = 'magenta', label = 'efficient version on 128*128 image')
        plt.plot(x,dim_bruteforce, color = 'magenta', label = 'bruteforce version on 128*128 image', linestyle='dashed')
    elif dim == 256:
        plt.plot(x,dim_efficient, color = 'yellow', label = 'efficient version on 256*256 image')
        plt.plot(x,dim_bruteforce, color = 'yellow', label = 'bruteforce version on 256*256 image', linestyle='dashed')
    elif dim == 512:
        plt.plot(x,dim_efficient, color = 'black', label = 'efficient version on 512*512 image')
        plt.plot(x,dim_bruteforce, color = 'black', label = 'bruteforce version on 521*512 image', linestyle='dashed')
        
    dim*=2

plt.xlabel('window size')
plt.ylabel('time')
plt.show()
```

```python
# 1.4

def brute_force_median_filter(im, k):
    start = time.time()
    h, w = img.shape
    p = (k-1)//2
    new_img = np.zeros((h, w), img.dtype)
    for x in range(w):
        for y in range(h):
            vec = []
            for i in range(-p,p+1):
                for j in range(-p,p+1):
                    if x+i>=0 and x+i<w and y+j>=0 and y+j<h:
                        vec.append(img[x+i,y+j])
            new_img[x,y]=np.median(vec)
            
    end = time.time()
    return end-start,new_img

def median_filter(im, k):
    start = time.time()
    # Get the shape of the 4d array needed for filtering     
    view_shape = tuple(np.subtract(im.shape, (k,k)) + 1) + (k,k)
    strd = np.lib.stride_tricks.as_strided
    # Get the new view of the array as required for filtering     
    subM = strd(im, shape = view_shape, strides = im.strides * 2)
    # for every i,j element in filter with <k,l> 2d array of subM
    # and return thier sum of as an i,j 2d array   
    end = time.time()
    return end-start, np.median(subM,axis=(2,3))

img = Image.open('../images/ceramic.jpg').convert('L')
img = np.array(img)

for k in range(3,14,2):
    t, op_img = median_filter(img,k)
    plt.imshow(op_img,cmap='gray')
    plt.title('grayscale ceramic image with {}*{} optimised median filter'.format(k,k))
    plt.xticks([]), plt.yticks([]) 
    plt.show()
```

```python
# 1.5
# As the brainNoisy.jpg image contains salt noise, so we should use min filter to remove it.
original_img = Image.open('../images/brainNoisy.jpg').convert('L')
original_img = np.array(original_img)
plt.imshow(original_img,cmap='gray')
plt.title('original image')
plt.xticks([]), plt.yticks([]) 
plt.show()

# we have to use filter of size 11 to remove most of the salt noise from the image
t, op_img = median_filter(original_img,11)
plt.imshow(op_img,cmap='gray')
plt.title('grayscale brain image with 11*11 optimised median filter')
plt.xticks([]), plt.yticks([]) 
plt.show()
```
